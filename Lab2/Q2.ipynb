```python
import numpy as np
import random
import matplotlib.pyplot as plt

X = np.array([random.uniform(1, 3) for _ in range(6)])
Y = np.array([1.7 + 0.5 * x for x in X])
```

```python
def gradient_descent(X, Y, learning_rate, num_iterations):
    m = len(X)
    theta0 = 0
    theta1 = 0
    cost_history = []
    theta_values = []

    for _ in range(num_iterations):
        Y_pred = theta0 + theta1 * X
        error = Y_pred - Y
        cost = np.sum(error ** 2) / m
        cost_history.append(cost)

        gradient_theta0 = np.sum(error) / m
        gradient_theta1 = np.sum(error * X) / m

        theta0 -= learning_rate * gradient_theta0
        theta1 -= learning_rate * gradient_theta1
        theta_values.append((theta0, theta1))

    return theta_values,cost_history

   

```

```python
learning_rate = 0.01
num_iterations = 3

theta_values,cost_history = gradient_descent(X, Y, learning_rate, num_iterations)

# Print cost function values
for iteration, cost in enumerate(cost_history):
    print(f"Iteration {iteration+1} - Cost: {cost:.4f}")
```

```python
# Plot the different regression lines for each iteration
plt.figure(figsize=(10, 6))
# plt.scatter(X, Y, label='Data points')

for i, (theta0, theta1) in enumerate(theta_values):
    y_pred = theta0 + theta1 * X
    plt.plot(X, y_pred, label=f'Iteration {i+1}')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regression Lines')
plt.legend()
plt.grid(True)
plt.show()
```
